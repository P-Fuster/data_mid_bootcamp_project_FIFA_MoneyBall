Data Cleaning

```python
#Since ID is a unique value for each player we set is as the index
df = df.set_index('ID')
```

```python
df1 =df.copy()
df1 = df.drop(['Player Photo', 'Club Logo', 'Flag Photo'], axis = 1)
#I am dropping those columns, as to me the have no useful information whatsoever
```

```python
df1[df1.columns[df1.isna().any()]].isna().sum()
#Here we check the null value and see that Loan Date End almost all of it are nulls
#Therefore we will drop the whole column while only the nulls in the other cases
```

```python
#Here we drop the columns with almost all the null values
#We also drop the null values from our datafram
df1 = df1.drop(['Loan Date End'], axis = 1)
df1 = df1.dropna()
df1.shape
```

```python
#I saw that height and weight were objects, need to change format to kg and cm and into integer or floats
def height(x):
    ft = str(x).replace('"', '').replace('[', '').split('\'')[0]
    inch = str(x).replace('"', '').replace('[', '').split('\'')[1]
    if not ft:
        ft = 0
    else:
        ft = int(ft)
    if not inch:
        inch = 0
    else:
        inch = int(inch)
    return ft*30.38+inch*2.54

df1['Height'] = df1['Height'].apply(height)

df1['Weight'] = df1['Weight'].apply(lambda x: int(str(x).replace('lbs','')) * 0.4536)

#df[['Height', 'Weight']]
```

```python
#Change the format of value since some M other have k other 
df1['Value'] = df1['Value'].apply(lambda x: int(str(x).replace('€', '')
                                .replace('M','00000').replace(',','')
                                .replace('K', '000').replace('.','')))

df1['Value']
```

```python
df1['Wage'] = df1['Wage'].apply(lambda x: int(str(x).replace('€', '')
                                .replace('M','00000').replace(',','')
                                .replace('K', '000').replace('.','')))

df1['Wage']
```

```python
df1['IR'] = df1['IR'].apply(lambda x: int(str(x).replace('★', '')))
df1.IR
```

```python
#By looking at Trinidad&Tobago we seemed to have a problem with the &amp; we take care of it.
df1['Nationality'] = df1['Nationality'].apply(lambda x: str(x).replace('amp;',''))
df1['Nationality'].unique()
```

```python
#In growth we seem to have a lot of 0 values (5229 to be exact) and I dont even know what it is. Drop it
df1['Growth'].value_counts()
df1 = df1.drop(['Growth'], axis=1)
```

Feature Engineering

```python
# (X-y split comes first, or else your features will carry information from the target)
y = df['Wage']
X = df.drop(['Wage'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = np.object)
```

```python
import pickle
# Normalizing data
transformer = Normalizer() # Normalizer works column wise, so don't use it on the target 
transformer.fit(X_num) # you'll always only fit on the trainning data

# When Normalizing with encoded categorical data, you'll lose some information on the categories.
# When Normalizing only numericals, you'll be losing proportion between numericals and encoded categories.

with open('transformer.pkl', 'wb') as file:
    pickle.dump(transformer, file)

with open('transformer.pkl', 'rb') as file:
    transformer = pickle.load(file)

x_normalized = transformer.transform(X_num)
#pd.DataFrame(x_normalized)
x_normalized.shape
```

```python
cat_data = pd.get_dummies(X_cat, drop_first=True) # if you need to save the encoding info, this won't do
```

```python
X = np.concatenate([x_normalized, cat_data], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression() # y = a*x + b
model.fit(X_train,y_train)
```

```python
predictions  = model.predict(X_test)
```

```python
r2_score(y_test, predictions), mean_absolute_error(y_test, predictions), mean_squared_error(y_test, predictions, squared=False)
```

```python
rmse = mean_squared_error(y_test, predictions, squared=False)
mae = mean_absolute_error(y_test, predictions)

print("R2_score:", round(model.score(X,y),2))
print("RMSE:", rmse)
print("MAE:", mae)
```
